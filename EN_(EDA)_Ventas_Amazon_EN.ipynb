```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Use the .read_csv command to import the .csv file.
amazon = pd.read_csv(r'amazon.csv')
```

**Data Cleaning**

```python
# Check for any strange values in the dataframe by analyzing descriptive statistics
amazon.describe()
```

```python
# Display data types for each column and the count of non-null values
amazon.info()
```

The columns are currently defined as objects, which is inappropriate for numerical data such as discounted_price, actual_price, discount_percentage, rating, and rating_count. Converting these columns to float64 caused an error due to symbols like commas, currency signs, and percentages. These symbols must be removed before changing the data type.

**Feature Engineering**

Remove symbols from columns and convert them to float64.

**Conversion of discounted_price and actual_price Columns**

```python
# Remove commas and currency signs from `discounted_price` and `actual_price` columns using the .str.replace() command
amazon['discounted_price'] = amazon['discounted_price'].str.replace("₹", '').str.replace(",", '').astype('float64')
amazon['actual_price'] = amazon['actual_price'].str.replace("₹", '').str.replace(",", '').astype('float64')
```

Both discounted and actual prices are in Indian Rupees.

Convert Indian Rupees to Euros. The exchange rate from Rupees to Euros is 0.012.

```python
# Apply a function to each row using the .apply() command
exchange_rate = 0.012
amazon['discounted_price'] = amazon['discounted_price'].apply(lambda x: x * exchange_rate)
amazon['actual_price'] = amazon['actual_price'].apply(lambda x: x * exchange_rate)

# Round prices to 2 decimal places using the round() command
amazon['discounted_price'] = round(amazon['discounted_price'], 2)
amazon['actual_price'] = round(amazon['actual_price'], 2)
```

**Creating the discount_amount Column**

It is useful to know both the exact value of discounts and the corresponding percentage.

```python
amazon['discount_amount'] = amazon['actual_price'] - amazon['discounted_price']
```

**Conversion of the discount_percentage Column**

```python
# Remove the % symbol from the `discount_percentage` column using the .str.replace() command
amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%', '').astype('float64')

# Divide `discount_percentage` by 100 to convert the values to decimals
amazon['discount_percentage'] = amazon['discount_percentage'] / 100
```

**Conversion of the rating Column**

Although the rating column appears correct, there might be some incompatible value with the float64 data type, causing an error when trying to change the data type to float64.

```python
# Display unique values in a column and their counts using the .value_counts() command
amazon['rating'].value_counts()
```

In the lower part of the list, there is a symbol “|” blocking the data conversion to float64. To resolve this, identify the affected row in the rating column, check the correct rating on the product page, and replace the “|” symbol with the appropriate value.

```python
# Select the row(s) where the rating value is "|" using the .loc[] command
amazon.loc[amazon['rating'] == '|']
```

The product page indicates that the average rating of the product is 3.9.

```python
# Replace "|" with 3.9 in the rating column using the .str.replace() command
amazon['rating'] = amazon['rating'].str.replace('|', '3.9').astype('float64')
```

**Conversion of the rating_count Column**

```python
# Remove commas from the `rating_count` column using the .str.replace() command
amazon['rating_count'] = amazon['rating_count'].str.replace(',', '').astype('float64')
```

**Handling Missing Values**

Remove null values.

To ensure data integrity, it is necessary to remove or replace null values with an average or median value.

```python
# Display the number of null values in each column using the .isna() and .sum() commands
amazon.isna().sum()
```

There are 2 null values in the rating_count column.

```python
# Find the index number of rows with 1 or more null values using the .index command
amazon.iloc[amazon[(amazon.isna().sum(axis=1) >= 1)].index]
```

Replace null values with the mean of rating_count.

```python
# Replace null values with the rounded mean value using the .fillna() command
amazon['rating_count'] = amazon['rating_count'].fillna(round(amazon["rating_count"].mean()))

# Display the number of null values in each column using the .isna() and .sum() commands
amazon.isna().sum()
```

**Removing Duplicate Values**

```python
# Remove duplicate rows using the .drop_duplicates() command
amazon = amazon.drop_duplicates()
```

Remove whitespace from product_id.

```python
# Use the .str.strip() command to remove any leading and trailing whitespace
amazon['product_id'] = amazon['product_id'].str.strip()
```

**Splitting Multiple Categories into Separate Columns**

It is necessary to split each category into individual columns and adjust the names to be clearer and more readable. The categories in the corresponding column are separated by the “|” symbol (e.g., Computers & Accessories | Accessories & Peripherals). To make these data more manageable for analysis.

```python
# Use the .str.split() command to split string values in the category column using "|" as a delimiter
splitcategory = amazon['category'].str.split('|', expand=True)
splitcategory

# Rename columns 0 to "category" and 1 to "subcategory" using the .rename() command
splitcategory = splitcategory.rename(columns={0: 'category', 1: 'subcategory'})
splitcategory
```

```python
# Format category values to be more readable using .str.replace()
splitcategory['category'] = splitcategory['category'].str.replace('&', ' & ')
splitcategory['subcategory'] = splitcategory['subcategory'].str.replace('&', ' & ')

# Drop the category column from the amazon dataframe using the .drop() command
amazon = amazon.drop(columns='category')

# Add the new category and subcategory columns to the amazon dataframe
amazon['category'] = splitcategory['category']
amazon['subcategory'] = splitcategory['subcategory']
```

**Creating a Column to Classify Products Based on Rating Scores**

Each numerical range corresponds to a specific classification (e.g., 0 - 0.9 stars = Very Poor to 5 stars = Excellent). The classification column will be based on product ratings.

```python
# Create a for loop and an if-else statement to categorize the rating scale
ranking = []

for score in amazon['rating']:
    if score <= 0.9:
        ranking.append('Very Poor')
    elif score <= 1.9:
        ranking.append('Poor')
    elif score <= 2.9:
        ranking.append('Average')
    elif score <= 3.9:
        ranking.append('Good')
    elif score <= 4.9:
        ranking.append('Very Good')
    elif score == 5.0:
        ranking.append('Excellent')

# Add the ranking list as a ranking column in the amazon dataframe
amazon['ranking'] = ranking
amazon['ranking'] = amazon['ranking'].astype('category')
```

**Creating a Dataframe with Columns for Analysis**

A new dataframe has been created that only retains the essential columns for analysis, after cleaning the data. Some columns, such as img_link and product_link, will not be included in the analysis.

```python
amazon_clean = amazon[['product_id', 'product_name', 'actual_price', 'discounted_price', 'discount_amount', 'discount_percentage', 'category', 'subcategory', 'rating', 'rating_count', 'ranking']]
amazon_clean
```

**Creating a Dataframe with Reviewer Data**

To allow a more detailed analysis of reviewers

, an additional structure is needed, as the amazon_clean dataframe focuses on products sold by Amazon. An additional dataframe including reviewer data is necessary to analyze customers who reviewed the products. Currently, the user_name and user_id columns contain multiple values separated by commas; these values must be split into individual rows instead of columns.

```python
# Use the .str.split() command to split string values in the user_id and user_name columns using ',' as a delimiter
split_user_id = amazon['user_id'].str.split(',', expand=False)
split_user_name = amazon['user_name'].str.split(',', expand=False)

# Use the .explode() command to split each list element into a row
id_rows = split_user_id.explode()
name_rows = split_user_name.explode()

# Use the DataFrame() command to create a dataframe using the exploded lists
df_id_rows = pd.DataFrame(id_rows)
df_name_rows = pd.DataFrame(name_rows)

# Add product_name, category, and subcategory columns from the amazon_clean dataframe to the df_name_rows dataframe
df_name_rows['product_name'] = amazon_clean['product_name']
df_name_rows['category'] = amazon_clean['category']
df_name_rows['subcategory'] = amazon_clean['subcategory']

# Reset the index so that each row has its own index number using the .reset_index() command
df_id_rows = df_id_rows.reset_index(drop=True)
df_name_rows = df_name_rows.reset_index(drop=True)

# Use the .merge() command to merge two dataframes together
reviewers = pd.merge(df_id_rows, df_name_rows, left_index=True, right_index=True)
reviewers
```

```python
# Display the number of null values in the reviewers dataframe using the .isna() and .sum() commands
reviewers.isna().sum()
```

The reviewers dataframe now contains a single user name entry per row, thanks to the application of the .explode() command, which allowed each username to occupy a separate row. Originally, in the Amazon dataframe, the user_name column included several usernames in a single row. As a result, the reviewers dataframe has 11,503 rows, in contrast to the 1,465 rows in the original Amazon dataframe.

**Creating a Correlation Matrix to Determine How Variables are Related**

Variables that strongly influence each other will have a correlation close to 1 (positive or negative). Variables that do not have much influence on each other will have a correlation close to 0. The correlation matrix is a way to easily see the strength of the relationship between 2 variables.

```python
# Set the size of the heatmap using the .set() command
sns.set(rc={'figure.figsize': (9, 6)})

# Create a dataframe to use in the correlation matrix
# The dataframe should only have variables with numerical data types
amazon_correl = amazon_clean[['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']]

# Create a correlation matrix using the .corr() command
# The correlation matrix will be text only
# Use the .heatmap() command to create a heatmap of the correlation matrix, making correlation values easier to visualize
sns.heatmap(amazon_correl.corr(), linewidths=1, annot=True, annot_kws={'size': 10}, vmax=1, vmin=-1, center=0, cmap='PiYG')

plt.title('Correlation Matrix', fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

**Finding the Percentage of Different Products by Category and Subcategory**

As with finding the number of different products by category and subcategory, finding the percentage will provide insight into customer interest by category and subcategory. Percentages are easier to visualize when comparing how each category contributes to the whole.

```python
# Create a dataframe containing the data to be used in a pivot table
# Create a percentage column where all values are 1
cat_pivot = amazon_clean.loc[:, ['category', 'subcategory', 'product_id']]
cat_pivot['percentage'] = 1

# Group results by category using the .groupby() command
# Create a function using lambda to create a pivot table and calculate the total percentage for each category and subcategory
# Create a pivot table using the .pivot_table() command
# The pivot table will group by categories and subcategories and sum the 1's in the percentage column to calculate the percentages for each subcategory
cat_pivot.groupby('category').apply(lambda cat_pivot_table: cat_pivot_table.pivot_table(index=['subcategory'], values=['percentage'], aggfunc=lambda x: round(sum(x)/sum(cat_pivot['percentage']) * 100, 2), margins=True))
```

The “Electronics”, “Computers & Accessories”, and “Home & Kitchen” categories constitute 97% of the product variety. Customers who buy electronics mainly buy home theaters, videos, televisions, mobile phones, and accessories. Customers who buy computers and accessories mainly buy accessories and peripherals. Customers who buy “Home & Kitchen” mainly buy kitchen and home appliances.

**Finding the Average Price and Discount Price by Category**

To understand how much revenue each category generates, the data must also include a time frame and the number of units sold. However, finding the average price per product and the number of different products sold will provide an indication of the total revenue generated by each category.

```python
# Create an order list so that the bars in the bar chart are arranged in descending order
# Group values based on a column using the .groupby() command
# Get the average of each group using the .mean() command
# Sort average values in descending order using the .sort_values() command
cat_price_order = amazon_clean.groupby('category')['actual_price'].mean().sort_values(ascending=False).index.values
cat_discount_order = amazon_clean.groupby('category')['discounted_price'].mean().sort_values(ascending=False).index.values

fig, ax = plt.subplots(2, 1, figsize=(15, 15))

sns.barplot(ax=ax[0], data=amazon_clean, x='actual_price', y='category', errorbar=None, order=cat_price_order).set(title='Average Price by Category', xlabel='Average Price', ylabel='Category')
sns.barplot(ax=ax[1], data=amazon_clean, x='discounted_price', y='category', errorbar=None, order=cat_discount_order).set(title='Average Discount Price by Category', xlabel='Average Discount Price', ylabel='Category')

ax[0].bar_label(ax[0].containers[0])
ax[1].bar_label(ax[1].containers[0])

plt.show()
```

The “Electronics” category has the highest average price per product before and after discounts. The second and third highest average prices per product before discounts are the “Home & Kitchen” and “Car & Motorbike” categories. On average, both categories have similar product prices. After applying discounts, the “Home & Kitchen” and “Car & Motorbike” categories switch places. The “Home & Kitchen” category may have higher average discounts compared to the “Cars & Motorbikes” category.

Although the “Computers & Accessories” category has the second highest number of unique products, the average price per product both before and after discounts is low compared to “Electronics” and “Home & Kitchen” products. The low prices of “Computers & Accessories” products may contribute to the wide variety of products sold in that category.

**Checking for heteroscedasticity.**

Residuals are the difference between the predicted value of the dependent variable (discounted price) and the actual value of the dependent variable. Heteroscedasticity occurs when the variance of the residuals is not constant as the value of the dependent variable changes. Heteroscedasticity suggests that the model may need additional independent variables to better predict the dependent variable. When plotting the predicted values and residuals on a scatter plot, a model with heteroscedasticity will have data points that spread out.

```python
# Import the statsmodels library.

from statsmodels.formula.api import ols

# Create a linear regression model.
# Use the ols() command to perform ordinary least squares linear regression.

model = ols(formula='discounted_price ~ actual_price',
            data=amazon_correl[['discounted_price', 'actual_price']]).fit()

# Create a dataframe with the independent variable.

X_OLS = amazon_correl['actual_price']

# Use the .predict command to predict the values of the dependent variable (discounted price).

fitted_values = model.predict(X_OLS)

# Use the .resid command to calculate the difference between the predicted values and the actual values.

residuals = model.resid
```

```python
# Use a scatter plot to graph the fitted values vs. the residuals.
# If the data has heteroscedasticity, the data points will spread outward.

plt.scatter(fitted_values, residuals)

plt.xlabel('Fitted Values')
plt.ylabel('Residuals')

plt.show()
```

**The model has heteroscedasticity.** As the fitted values increase, the spread of the points also increases. This indicates that **when the model predicts the discounted price of a product, the higher the predicted discounted price, the more prone to errors the model becomes.** Heteroscedasticity suggests that other variables are needed to predict the discounted price. However, other variables, such as discount percentage, rating, and rating count, have very little correlation and impact on the discounted price. There may be other factors not included in the dataset that could influence discounted prices.

**Checking for residual normality.**

**In this analysis, we will evaluate the presence of heteroscedasticity in our linear regression model** that predicts discounted_price based on actual_price.

- We will fit a simple linear regression model.
- We will perform the Breusch-Pagan test to detect heteroscedasticity in the model.
- We will generate several plots to visualize the residuals and assess the presence of heteroscedasticity.

```python
import seaborn as sns
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.formula.api import ols

# Fit the regression model
model = ols('discounted_price ~ actual_price', data=amazon).fit()

# Perform the Breusch-Pagan test
test = het_breuschpagan(model.resid, model.model.exog)
labels = ['LM Stat', 'LM Test p-value', 'F-Stat', 'F-Test p-value']
breusch_pagan_results = dict(zip(labels, test))

# Plot the residuals
residuals = model.resid
fitted = model.fittedvalues

# Residuals vs Fitted values plot
plt.figure(figsize=(10, 6))
sns.residplot(x=fitted, y=residuals, lowess=True, line_kws={'color': 'red'})
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()

# Histogram of residuals
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.xlabel('Residual Values')
plt.title('Distribution of Residuals')
plt.show()

# QQ plot of residuals
sm.qqplot(residuals, line='45')
plt.title('QQ Plot of Residuals')
plt.show()

# Show the results of the Breusch-Pagan test
breusch_pagan_results
```

**Residuals vs Fitted Values:**

The plot shows an increasing spread of residuals as the fitted values increase.
The fitted red line suggests an upward trend, which indicates the presence of heteroscedasticity, as the residuals do not have a constant variance.

**Histogram of Residuals:**

The histogram shows a distribution of residuals with a high concentration around zero.
The presence of long tails at both ends indicates possible outliers and deviations from normality.

**QQ Plot of Residuals:**

The QQ plot shows a significant deviation of the points from the reference line, especially at the extremes.
This suggests that the residuals do not follow a normal distribution and there are outliers present.

To consider the use of robust standard errors and obtain more reliable inferences, **I will fit the linear regression model with robust standard errors.**
We will use **cov_type='HC3'** to indicate that we are using the HC3 type heteroscedasticity correction. This adjusts the model's standard errors to be robust to heteroscedasticity.

```python
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.diagnostic import het_breuschpagan, het_white
import statsmodels.api as sm

np.random.seed(0)
X = np.random.rand(100)
y = 2 * X + np.random.rand(100) * 0.5
data = pd.DataFrame({'X': X, 'y': y})

# Fit the linear regression model
model = smf.ols('y ~ X', data=data).fit()

# Fit the model with robust standard errors
model_robust = model.get_robustcov_results(cov_type='HC3')

# Show the summary of the model with robust standard errors
print(model_robust.summary())

# Calculate the VIF for each predictor variable
X = data[['X']]  # If you have more variables, add them here
X = sm.add_constant(X)
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Show the VIF values
print(vif_data)

# Breusch-Pagan test
bp_test = sm.stats.diagnostic.het_breuschpagan(model.resid, model.model.exog)
bp_test_results = {
    'Lagrange multiplier statistic': bp_test[0],
    'p-value': bp_test[1],
    'f-value': bp_test[2],
    'f p-value': bp_test[3]
}
print("\nBreusch-Pagan test results:")
print(bp_test_results)

# White test
white_test = sm.stats.diagnostic.het_white(model.resid, model.model.exog)
white_test_results = {
    'Test Statistic': white_test[0],
    'Test Statistic p-value': white_test[1],
    'F-Statistic': white_test[2],
    'F-Statistic p-value': white_test[3]
}
print("\nWhite test results:")
print(white_test_results)
```

The lack of significant autocorrelation in the residuals suggests that the regression model has adequately captured the data structure and that the residuals are essentially random noise. **This indicates that the model is appropriate and that the model parameter estimates are reliable.**

```python
import statsmodels.api as sm

# Define the dependent and independent variables
X = amazon_clean['actual_price']
Y = amazon_clean['discounted_price']

# Add a constant (intercept) to the model
X = sm.add_constant(X)

# Fit the robust model
model_robust = sm.RLM(Y, X, M=sm.robust.norms.HuberT())
results = model_robust.fit()

# Obtain the model coefficients
intercept = results.params['const']
slope = results.params['actual_price']

# Format the equation
equation = f"Y = {slope:.3f}X + {intercept:.3f}"
interpretation = f"This indicates that for every 1 Euro increase in the actual price, the discounted price increases by {slope:.3f} Euros."

# Show the results
print(f"The discounted price (Y) based on the actual price (X) results in the equation {equation}. {interpretation}")
```
