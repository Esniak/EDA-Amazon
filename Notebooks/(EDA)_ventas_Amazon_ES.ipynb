Estas columnas deberían ser del tipo **float64**. Al intentar realizar la conversión de estas columnas a **float64**, se produjo un **error** debido a la presencia de **símbolos** como comas, signos de moneda y porcentajes. Para solucionar este problema, es necesario **eliminar estos símbolos** antes de cambiar el tipo de dato.




**"Feature engineering"**

**Eliminar símbolos de las columnas y convertirlas en el tipo de datos float64.**

- **Conversion de las columnas discounted_price y actual_price**

````python
# para eliminar las comas y ₹ de las columnas discounted_price y actual_price, utilizo  el comando .str.replace()

amazon['discounted_price'] = amazon['discounted_price'].str.replace("₹",'').str.replace(",",'').astype('float64')

# para convertir el tipo de dato en las columnas discounted_price y actual_price a float64, utilizo el comando .astype()

amazon['actual_price'] = amazon['actual_price'].str.replace("₹",'').str.replace(",",'').astype('float64')
````

Tanto los precios de descuento como los reales están en Rupias.

 **Convierto las Rupias indias a Euros**. El tipo de cambio de Rupias a Euros es de **0,012.**

````python
# para aplicar una función a cada fila, utilizo el comando .apply()
# utilizo lambda para crear una función.

exchange_rate = 0.012
amazon['discounted_price'] = amazon['discounted_price'].apply(lambda x: x * exchange_rate)
amazon['actual_price'] = amazon['actual_price'].apply(lambda x: x * exchange_rate)

# para redondear los precios a 2 decimales, utilizo el comando round()

amazon['discounted_price'] = round(amazon['discounted_price'], 2)
amazon['actual_price'] = round(amazon['actual_price'], 2)
````

- **Creación de la columna discount_amount.**

Es útil conocer tanto el **valor exacto de los descuentos** como el **porcentaje** correspondiente.

````python
amazon['discount_amount'] = amazon['actual_price'] - amazon['discounted_price']
````

- **Conversión de la columna discount_percentage.**

````python
# para eliminar el % de la columna discount_percentage, utilizo el comando .str.replace()
# para convertir el tipo de dato en la columna discount_percentage a float64, utilizo el comando .astype()

amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%','').astype('float64')

# Divido discount_percentage por 100 para que los valores estén en decimales.

amazon['discount_percentage'] = amazon['discount_percentage'] / 100
````

- **Conversión de la columna rating.**

Aunque la **columna de calificación** parece correcta. Podría haber algún **valor incompatible** con el tipo de datos **float64**, lo que causa un **error** al intentar cambiar el tipo de datos a **float64**.

Para resolver este problema, identificaré la **fila afectada** en la columna de **calificación**, verificaré la **calificación correcta** en la **página web del producto**, y sustituiré el **símbolo "|"** por el valor adecuado.


````python
# para seleccionar la(s) fila(s) donde el valor de rating = |, utilizo el comando .loc[]

amazon.loc[amazon['rating'] == '|']
````

**Creo una columna que clasifique los productos en función de la puntuación de calificación.**

Cada rango numérico corresponderá a una **clasificación específica** (medidas de: **0 - 0.9 estrellas = Muy pobre** a **5 estrellas = Excelente**). La **columna de clasificación** se basará en las **calificaciones de los productos**.



````python
# Creo un bucle for y una declaración if else para categorizar la escala de calificación.
# utilizo el comando .append() para agregar categorías de clasificación a una lista de clasificación.

ranking = []

for score in amazon['rating']:
    if score <= 0.9: ranking.append('Muy Malo')
    elif score <= 1.9: ranking.append('Malo')
    elif score <= 2.9: ranking.append('Promedio')
    elif score <= 3.9: ranking.append('Bueno')
    elif score <= 4.9: ranking.append('Muy Bueno')
    elif score == 5.0: ranking.append('Excelente')

# Agrego la lista de clasificación como una columna de clasificación en el dataframe de amazon.
# utilizo el comando .astype() para convertir los valores en la columna de clasificación en un tipo de dato de categoría.

amazon['ranking'] = ranking

amazon['ranking'] = amazon['ranking'].astype('category')
````

**Creo un dataframe con reviewer data.**

Para permitir un análisis más detallado de los **revisores**, se necesita una estructura adicional, ya que el **dataframe amazon_clean** se enfoca en los productos vendidos por Amazon. Es necesario un **dataframe adicional** que incluya los datos de los revisores para analizar a los **clientes que revisaron los productos**. Actualmente, las columnas **user_name** y **user_id** contienen múltiples valores separados por comas; estos valores deben dividirse en **filas individuales** en lugar de columnas.



````python
# utilizo el comando .str.split() para dividir los valores de cadena en las columnas user_id y user_name usando ',' como delimitador.
# expand = False insertará cada elemento dividido en una lista.

split_user_id = amazon['user_id'].str.split(',', expand = False)
split_user_name = amazon['user_name'].str.split(',', expand = False)
````

````python
# utilizo el comando .explode() para dividir cada elemento de una lista en una fila.
# Nota: Aunque cada elemento de la lista está en una fila diferente, los elementos comparten el mismo número de índice.
# Ejemplo: Una lista con 5 valores se dividirá en 5 filas, pero cada una de esas 5 filas tendrá el mismo número de índice.

id_rows = split_user_id.explode()
name_rows = split_user_name.explode()
````

````python
# utilizo el comando DataFrame() para crear un dataframe usando the exploded lists.

df_id_rows = pd.DataFrame(id_rows)
df_name_rows = pd.DataFrame(name_rows)
````

````python
# Agrego las columnas product_name, category y subcategory del dataframe amazon_clean al dataframe df_name_rows.

df_name_rows['product_name'] = amazon_clean['product_name']
df_name_rows['category'] = amazon_clean['category']
df_name_rows['subcategory'] = amazon_clean['subcategory']
````

````python
# utilizo el comando .reset_index() para restablecer el índice de modo que cada fila tenga su propio número de índice.

df_id_rows = df_id_rows.reset_index(drop = True)
df_name_rows = df_name_rows.reset_index(drop = True)
````

**Encontrar el porcentaje de diferentes productos por categoría y subcategoría.**

Al igual que encontrar el número de productos diferentes por categoría y subcategoría, encontrar el porcentaje proporcionará información sobre el interés de los clientes por categoría y subcategoría. Los porcentajes son más fáciles de visualizar cuando se compara cómo cada categoría contribuye al conjunto.

````python
# creo un dataframe que contiene los datos que se utilizarán en una tabla dinámica.
# creo una columna de porcentaje donde todos los valores son 1.

cat_pivot = amazon_clean.loc[:, ['category', 'subcategory', 'product_id']]
cat_pivot['percentage'] = 1

# Utilizo el comando groupby() para agrupar los resultados por categoría.
# Utilizo el comando lambda para crear una función que creará una tabla dinámica y calculará el total de porcentaje para cada categoría y subcategoría.
# Utilizo el comando .pivot_table() para crear una tabla dinámica.
# La tabla dinámica agrupará por categorías y subcategorías y sumará los 1 en la columna de porcentaje para calcular los porcentajes para cada subcategoría.

cat_pivot.groupby('category').apply(lambda cat_pivot_table: cat_pivot_table.pivot_table(index = ['subcategory'], values = ['percentage'],
                                                                                        aggfunc = lambda x: round(sum(x)/sum(cat_pivot['percentage']) * 100, 2),
                                                                                        margins = True))
````

Las categorías **"Electrónica", "Ordenadores y accesorios" y "Hogar y cocina" contituyen el 97 % de la variedad de productos.** Los clientes que compran electrónica compran principalmente cines en casa, vídeos, televisores, teléfonos móviles y accesorios. **Comprobación de la sobreajuste mediante la validación cruzada.**

La validación cruzada es otra forma de comprobar si el R2 es el resultado de un sobreajuste. La validación cruzada es el proceso de dividir el conjunto de datos original en k número de conjuntos iguales. **Comprobando si hay heteroscedasticidad.**

Los residuos son la diferencia entre el valor previsto de la variable dependiente (precio con descuento) y el valor real de la variable dependiente. La heteroscedasticidad se produce cuando la varianza de los residuos no es constante como el valor del cambio de la variable dependiente. La heteroescasticidad sugiere que el modelo puede necesitar variables independientes adicionales para predecir mejor la variable dependiente. Al trazar los valores predichos y los residuos en un gráfico de dispersión, un modelo con heteroescasticidad tendrá puntos de datos que se extienden.

````python
# Importar la biblioteca statsmodels.

from statsmodels.formula.api import ols

# Creo un modelo de regresión lineal.
# Utilizo el comando ols() para realizar una regresión lineal por mínimos cuadrados ordinarios.

model = ols(formula = 'discounted_price ~ actual_price',
          data = amazon_correl[['discounted_price', 'actual_price']]).fit()

# Creo un dataframe con la variable independiente.

X_OLS = amazon_correl['actual_price']

# Utilizo el comando .predict para predecir los valores de la variable dependiente (precio con descuento).

fitted_values = model.predict(X_OLS)

# Utilizo el comando .resid para calcular la diferencia entre los valores predichos y los valores reales.

residuals = model.resid
````

````python
# Utilizo un diagrama de dispersión para graficar los valores ajustados vs. los valores residuales.
# Si los datos tienen heterocedasticidad, los puntos de datos se dispersarán hacia afuera.

plt.scatter(fitted_values, residuals)

plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')

plt.show()
````

**El modelo tiene heteroscasticidad.** A medida que aumentan los valores ajustados, la diferencia de los puntos también aumenta. Esto indica que **cuando el modelo predice el precio con descuento de un producto, cuanto mayor sea el precio con descuento previsto, más propenso a errores se vuelve el modelo.** La heteroscedasticidad sugiere que se necesitan otras variables para predecir el precio con descuento. Sin embargo, otras variables, como el porcentaje de descuento, la calificación y el recuento de calificaciones, tienen muy poca correlación e impacto en el precio de descuento. Puede haber otros factores no incluidos en el conjunto de datos que puedan influir en los precios de descuento.

**Comprobación de la normalidad residual.**

Los residuos son la diferencia entre el valor predicho de la variable dependiente (precio con descuento) y el valor real de la variable dependiente. La normalidad es la condición de que un conjunto de datos tenga una distribución normal. La violación de la normalidad de los residuos tiene pocas consecuencias en los conjuntos de datos que son lo suficientemente grandes (generalmente, más de 200 puntos de datos).Sin embargo, la normalidad de los residuos tendrá un efecto al calcular otros valores estadísticos, como el error estándar o los valores P

````python
# Utilizo un histograma para determinar si los residuos tienen una distribución normal.

fig, ax = plt.subplots(2, 1, figsize=(15, 15))

sns.histplot(residuos,
             kde = True,
             ax = ax[0]).set_xlabel('Valores Residuales')

import statsmodels.api as sm

sm.qqplot(residuos,
          line = 's',
          ax = ax[1])

plt.show()
````

**Los gráficos muestran que los residuos no tienen una distribución normal.** Para que los residuos tengan una distribución normal, los puntos de datos en el gráfico QQ deben seguir la línea de cerca.

````python
import seaborn as sns
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.formula.api import ols

# Ajustar el modelo de regresión
model = ols('discounted_price ~ actual_price', data=amazon).fit()

# Realizar la prueba de Breusch-Pagan
test = het_breuschpagan(model.resid, model.model.exog)
labels = ['LM Stat', 'LM Test p-value', 'F-Stat', 'F-Test p-value']
breusch_pagan_results = dict(zip(labels, test))

# Graficar los residuos
residuals = model.resid
fitted = model.fittedvalues

# Gráfico de residuos vs valores ajustados
plt.figure(figsize=(10, 6))
sns.residplot(x=fitted, y=residuals, lowess=True, line_kws={'color': 'red'})
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')
plt.title('Residuos vs Valores Ajustados')
plt.show()

# Histograma de residuos
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.xlabel('Valores Residuales')
plt.title('Distribución de los Residuos')
plt.show()

# Gráfico QQ de residuos
sm.qqplot(residuals, line ='45')
plt.title('Gráfico QQ de los Residuos')
plt.show()

# Mostrar los resultados de la prueba de Breusch-Pagan
breusch_pagan_results
````

**Ajustar el modelo de regresión lineal con errores estándar robustos**

Para considerar el uso de errores estándar robustos y obtener inferencias más confiables, **voy ajustar el modelo de regresión lineal con errores estándar robustos.**  
usaremos **cov_type='HC3'** indica que estamos utilizando la corrección de heteroscedasticidad de tipo HC3. Esto ajusta los errores estándar del modelo para que sean robustos frente a la heteroscedasticidad.

````python
# Importar las librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.diagnostic import het_breuschpagan, het_white

# Paso 1: Importar el dataset
amazon_correl = pd.read_csv(r'/content/amazon_dataset_final.csv')

# Paso 2: Selección de características y variable objetivo (solo 'actual_price')
X = amazon_correl[['actual_price']]
Y = amazon_correl['discounted_price']

# Paso 3: Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print(f"\nNúmero de observaciones en entrenamiento: {X_train.shape[0]}")
print(f"Número de observaciones en prueba: {X_test.shape[0]}")

# Agregar una constante (intercepto) al modelo
X_train_const = sm.add_constant(X_train)
X_test_const = sm.add_constant(X_test)

# Paso 4: Calcular el VIF para cada variable predictora en el conjunto de entrenamiento
vif_data = pd.DataFrame()
vif_data["Feature"] = X_train_const.columns
vif_data["VIF"] = [
    variance_inflation_factor(X_train_const.values, i)
    for i in range(X_train_const.shape[1])
]

# Mostrar los valores de VIF
print("\nValores de VIF para cada variable predictora (Entrenamiento):")
print(vif_data)

# Paso 5: Ajustar el modelo de regresión lineal (OLS) en el conjunto de entrenamiento
model_ols = sm.OLS(Y_train, X_train_const).fit()

# Ajustar el modelo con errores estándar robustos (HC3)
model_robust = model_ols.get_robustcov_results(cov_type='HC3')

# Mostrar el resumen del modelo robusto
print("\nResumen del modelo de regresión lineal con errores estándar robustos (HC3) - Entrenamiento:")
print(model_robust.summary())

# Paso 6: Obtener los coeficientes del modelo de forma segura
params_series = pd.Series(model_robust.params, index=model_robust.model.exog_names)

print("\nParámetros del modelo robusto (Entrenamiento):")
print(params_series)

# Paso 7: Evaluar el modelo en el conjunto de prueba
y_pred_test = model_robust.predict(X_test_const)
y_true_test = Y_test

r_squared_test = r2_score(y_true_test, y_pred_test)
mse_test = mean_squared_error(y_true_test, y_pred_test)
rmse_test = np.sqrt(mse_test)
mae_test = mean_absolute_error(y_true_test, y_pred_test)
durbin_watson_stat_test = sm.stats.durbin_watson(model_robust.resid)

# Mostrar las métricas de evaluación en el conjunto de prueba
print("\nMétricas de Evaluación del Modelo en el Conjunto de Prueba:")
print(f"R²: {r_squared_test:.4f}")
print(f"MSE: {mse_test:.4f}")
print(f"RMSE: {rmse_test:.4f}")
print(f"MAE: {mae_test:.4f}")
print(f"Durbin-Watson: {durbin_watson_stat_test:.4f}")

# Paso 8: Test de Breusch-Pagan en el conjunto de entrenamiento
bp_test = het_breuschpagan(model_robust.resid, model_robust.model.exog)
bp_test_results = {
    'Lagrange multiplier statistic': bp_test[0],
    'p-value': bp_test[1],
    'f-value': bp_test[2],
    'f p-value': bp_test[3]
}
print("\nResultados de la prueba de Breusch-Pagan (Entrenamiento):")
for key, value in bp_test_results.items():
    print(f"{key}: {value}")

# Test de White en el conjunto de entrenamiento
white_test = het_white(model_robust.resid, model_robust.model.exog)
white_test_results = {
    'Test Statistic': white_test[0],
    'Test Statistic p-value': white_test[1],
    'F-Statistic': white_test[2],
    'F-Statistic p-value': white_test[3]
}
print("\nResultados de la prueba de White (Entrenamiento):")
for key, value in white_test_results.items():
    print(f"{key}: {value}")

# Paso 9: Obtener los coeficientes del modelo de forma segura
required_params = ['const', 'actual_price']
missing_params = [param for param in required_params if param not in params_series.index]

if not missing_params:
    intercept = params_series['const']
    slope = params_series['actual_price']

    # Formatear la ecuación del modelo
    equation = f"I = {slope:.3f}X + {intercept:.3f}"
    interpretation = f"Esto indica que por cada incremento de 1 Euro en el precio real, el precio con descuento aumenta en {slope:.3f} Euros."

    # Mostrar la ecuación del modelo
    print(f"\n{equation}. {interpretation}")
else:
    print(f"\nError: Los siguientes parámetros no se encontraron en el modelo: {missing_params}")
    print("Parámetros disponibles:")
    print(params_series)
    raise KeyError(f"Parámetros faltantes: {missing_params}")

# Paso 10: Guardar el modelo en disco utilizando pickle
filename = 'model.pkl'
try:
    with open(filename, 'wb') as file:
        pickle.dump(model_robust, file)
    print(f"\nModelo guardado en {filename}")
except Exception as e:
    print(f"\nError al guardar el modelo: {e}")
    raise
````
